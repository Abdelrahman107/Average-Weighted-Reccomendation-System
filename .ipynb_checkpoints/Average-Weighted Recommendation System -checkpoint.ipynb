```python
#import modules
import pandas as pd
import numpy as np
```

```python
# read datasets --> link https://www.kaggle.com/tmdb/tmdb-movie-metadata
movies_credit = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
movies_credit.head()
```

```python
#Print shapes
print(movies_credit.shape)
print(movies.shape)
```

```python
# Renaming Credits,Column Movie_id to id
movies_credit= movies_credit.rename(columns={'movie_id':'id'})
movies_credit.head()
```

```python
#Merging on ID
movies=movies.merge(movies_credit,on='id')
```

```python
movies.head()
```

```python
#Dropping unwanted Columns 
movies=movies.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
```

```python
movies.head()
```

```python
movies.info()
```

Average weighted Rating Equation  based on IMDB 
### Link: http://trailerpark.weebly.com/imdb-rating.html?source=post_page

 

```python
r = movies['vote_average']
v = movies['vote_count']
m = movies['vote_count'].quantile(0.70)
c = movies['vote_average'].mean()
```

```python
movies['weighted_rating'] = ((r*v)+ (c*m))/(v+m)
```

```python
movies.head()
```

```python
##Sort movies by weighted rating analogous to top 250IMDB
movie_sorted = movies.sort_values('weighted_rating',ascending=False)

```

```python
movie_sorted[['original_title', 'vote_count', 'vote_average', 'weighted_rating', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

graph = movies.sort_values('weighted_rating',ascending=False)
graph.head()
 
```

```python
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=graph['weighted_rating'].head(10), y=graph['original_title'].head(10),data=graph)
plt.title('Best Movies by Weighted Average votes', weight='bold')
plt.xlim(4, 10)
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

```python
# Taking Popularity into consideration --> Plotting it.
graph2 = movies.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
axis2=sns.barplot(x=graph2['popularity'].head(10), y=graph2['original_title'].head(10),data=graph2)
plt.title('Best Movies by Popularity', weight='bold')
plt.xlabel('Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

```python
#using MinMaxScaler to convert both columns into numbers from 0 to 1

from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
scaled_movie = scaling.fit_transform(movies[['weighted_rating','popularity']])
 
scaled_movie_df = pd.DataFrame(scaled_movie,columns=['weighted_rating_normalized','popularity_normalized'])
scaled_movie_df.head()
```

```python
movies[['weighted_rating_normalized','popularity_normalized']]=scaled_movie_df
movies.head()
```

```python
movies['score'] = movies['weighted_rating_normalized'] * 0.5 + movies['popularity_normalized'] * 0.5
movies_sorted_score= movies.sort_values(['score'], ascending=False)
movies_sorted_score[['original_title', 'weighted_rating_normalized', 'popularity_normalized', 'score']].head(20)
```

```python
#Plotting
plt.figure(figsize=(12,6))
axis2=sns.barplot(x=movies_sorted_score['score'].head(10), y=movies_sorted_score['original_title'].head(10),data=movies_sorted_score)
plt.title('Best Movies by score', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```
